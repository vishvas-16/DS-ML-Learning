### Vector Addition and Multiplication

**Goal:**  
Understand the basics of vector operations (addition, scalar multiplication, and dot product) with NumPy.

**Key Steps Covered:**
1. Representing vectors in NumPy
2. Performing vector addition
3. Performing scalar multiplication
4. Computing the dot product
5. Geometric interpretation (magnitude, orthogonality)



```python
import numpy as np

# Define two vectors
v1 = np.array([2, 3])
v2 = np.array([1, 4])

print("Vector v1:", v1)
print("Vector v2:", v2)
```

```python
# Vector addition
v_add = v1 + v2
print("v1 + v2 =", v_add)
```

```python
# Multiply vector by a scalar
scalar = 3
v_scalar = scalar * v1
print(f"{scalar} * v1 =", v_scalar)
```

```python
# Dot product of two vectors
dot_product = np.dot(v1, v2)
print("v1 · v2 =", dot_product)
```

```python
import matplotlib.pyplot as plt

# Plot vectors for visualization
origin = [0, 0]

plt.quiver(*origin, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(*origin, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
plt.quiver(*origin, v_add[0], v_add[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1+v2')

plt.xlim(0, 6)
plt.ylim(0, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
origin = [0, 0]

plt.quiver(*origin, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(*origin, v_scalar[0], v_scalar[1], angles='xy', scale_units='xy', scale=1, color='b', label='2 * v1')


plt.xlim(0, 6)
plt.ylim(0, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
