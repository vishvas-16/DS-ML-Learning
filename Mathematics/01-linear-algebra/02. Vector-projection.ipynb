## Relation between Cosine and Dot Product

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dot product can also be calculated by multiplying the magnitude of both the vectors with cosθ, where θ is the angle between them.
# r ⋅ s = |r| |s| cos θ
r = np.array([2, 3])
s = np.array([3, 4])
dot_product = np.dot(r, s)
print(dot_product)
r_magnitude = np.sqrt(np.dot(r,r))
s_magnitude = np.sqrt(np.dot(s,s))
cosine_rule = dot_product / (r_magnitude * s_magnitude)
print("cos(θ):", cosine_rule)
```

## Projections

### Scalar Projection
<p>Tells the magnitude/ length of the shadow of s projected onto r or any vector replaced by r in the formula.
</p>

```python
# In the formula r ⋅ s = |r| |s| cos θ, |s| cos θ represents the projection of s onto r
# |s| cos θ i.e. Scalar projection =  r ⋅ s / |r|

#Projection of s onto r
project_s_on_r = dot_product/r_magnitude
print(project_s_on_r)
origin = np.array([0, 0])

#Projection of s onto x-axis
project_on_axis = np.dot(s, [1,0]) / 1*1
print(project_on_axis)
```

### Vector Projection
<p>Represents the actual vector in the coordinate space
</p>

```python
# Formula = (r ⋅ s / |r| * |r|) * r

# Projecting s onto r
vector_project = (dot_product / (r_magnitude * r_magnitude)) * r
print(vector_project)
```

```python

# Plot r, s, and the projection of s onto r
origin = np.array([0, 0])

plt.figure(figsize=(7, 7))
plt.quiver(*origin, r[0], r[1], angles='xy', scale_units='xy', scale=1, color='r', label='r')
plt.quiver(*origin, s[0], s[1], angles='xy', scale_units='xy', scale=1, color='b', label='s')

# Vector projection of s onto r
r_magnitude_squared = np.dot(r, r)
vector_projection = (dot_product / r_magnitude_squared) * r

# Plot the projection vector (from origin)
plt.quiver(*origin, vector_projection[0], vector_projection[1], angles='xy', scale_units='xy', scale=1, color='g', label='proj_r(s)')

# Optionally, draw a dashed line from the tip of the projection to the tip of s
plt.plot([vector_projection[0], s[0]], [vector_projection[1], s[1]], 'k--', label='Error/Residual')

plt.xlim(0, max(r[0], s[0], vector_projection[0]) + 2)
plt.ylim(0, max(r[1], s[1], vector_projection[1]) + 2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Vector r, s, and Projection of s onto r")
plt.show()
```
