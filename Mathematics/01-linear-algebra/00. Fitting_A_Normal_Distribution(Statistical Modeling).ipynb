# Using Statictical Modeling to fit a Bell Curve data set.


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

## Example dataset 

```python
heights_cm = [
    174.7, 170.0, 175.8, 182.3, 169.2, 168.5, 182.8, 176.6, 166.1, 169.8,
    171.5, 159.7, 170.1, 161.8, 177.8, 182.2, 162.7, 168.9, 178.6, 175.7,
    174.9, 177.3, 177.1, 172.9, 165.2, 162.3, 169.1, 168.3, 159.4, 169.9,
    169.1, 180.2, 167.3, 170.0, 170.6, 179.6, 162.1, 165.6, 167.9, 171.2
]
# Make a histogram (this gives frequencies for each bin)
counts, bins, patches = plt.hist(heights_cm, bins=10, edgecolor="black", alpha=0.6)

plt.xlabel("Height (m)")
plt.ylabel("Frequency")
plt.title("Frequency Distribution of Heights (raw data)")

plt.show()
```

## Calculating the sample Mean and sample Standard Deviation

```python
sample_mean = np.mean(heights_cm)
print("Sample mean:", sample_mean)

sample_std = np.std(heights_cm, ddof=1)  # ddof=1 for sample standard deviation
print("Sample standard deviation:", sample_std)
```

## Supplimenting values into Normal Probability Density Function and Plotiing the fitted line.

```python
from scipy.stats import norm

# Generate x values (heights) for plotting the normal curve
x = np.linspace(min(heights_cm), max(heights_cm), 100)

# Compute the normal PDF using the sample mean and std
pdf = norm.pdf(x, loc=sample_mean, scale=sample_std)

# Plot the histogram again
plt.hist(heights_cm, bins=10, edgecolor="black", alpha=0.6, density=True)
plt.plot(x, pdf, 'r-', label='Normal PDF (fit)')
plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.title("Normal Distribution Fit to Heights")
plt.legend()
plt.show()
```
